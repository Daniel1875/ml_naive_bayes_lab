```python
import numpy as np
import pandas as pd
import os
import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import math
```

We will first read the data into two arrays, each containing a number of strings, each string being a file but with all punctuation and all stopwords removed.

```python
data_dir = "C:/Users/cuiji/Desktop/ML2020LAB/ml_naive_bayes_lab-main/movies_reviews/"

positive_reviews = []
negative_reviews = []
stopwords = []

# opens the stopwords file, and stores the data
file = open('stop_words.txt')
stopwords = file.readlines()
stopwords = [x.strip() for x in stopwords]

# makes the stopwords include useless characters (punctuation, etc.)
stopwords = stopwords + [',','.',';', ':', '!', "*", '?','<','>','/','"',"'","-","_","+","=","(",")","\n", ' ', '&']

# reads the positive reviews, but with all punctuation and stopwords removed
for filename in os.listdir(data_dir+'pos'):
    file = open(data_dir+'pos/'+filename)
    data = file.read()
    data = data.replace('\n', '')
    data = data.split(' ')
    review = ""
    for i in data:
        if i not in stopwords:
            review += i
            review += ' '
    positive_reviews.append(review)
    file.close()

# same thing for negative reviews
for filename in os.listdir(data_dir+'neg'):
    file = open(data_dir+'neg/'+filename)
    data = file.read()
    data = data.replace('\n', '')
    data = data.split(' ')
    review = ""
    for i in data:
        if i not in stopwords:
            review += i
            review += ' '
    negative_reviews.append(review)
    file.close()
```

```python
# store all reviews into lists
reviews = []
class_label = []
for r in positive_reviews:
    reviews.append([r, 1])
for r in negative_reviews:
    reviews.append([r, 0])

# shuffles the list
random.shuffle(reviews)

# creates training and test sets
training_size = math.floor(len(reviews)*0.8)
X = [review[0] for review in reviews]
Y = [review[1] for review in reviews]
X_train = X[:training_size]
Y_train = Y[:training_size]
X_test = X[training_size:]
Y_test = Y[training_size:]

# use CountVectorizer
cv = CountVectorizer()
X_train = cv.fit_transform(X_train)
X_test = cv.transform(X_test)
```

```python
# build the model and find the accuracy
nb = MultinomialNB().fit(X_train, Y_train)
predicted = nb.predict(X_train)
print("Accuracy for train:", accuracy_score(Y_train, predicted))
predicted = nb.predict(X_test)
print("Accuracy for test:", accuracy_score(Y_test, predicted))
```

```python
# reads in the new reviews
new_pos = []
new_neg = []
for filename in os.listdir(data_dir+'new'):
    if 'pos' in filename:
        file = open(data_dir+'new/'+filename)
        data = file.read()
        data = data.replace('\n', '')
        data = data.split(' ')
        review = ""
        for i in data:
            if i not in stopwords:
                review += i
                review += ' '
        new_pos.append(review)
        file.close()
    else:
        file = open(data_dir+'new/'+filename)
        data = file.read()
        data = data.replace('\n', '')
        data = data.split(' ')
        review = ""
        for i in data:
            if i not in stopwords:
                review += i
                review += ' '
        new_neg.append(review)
        file.close()

# stores new reviews
new_reviews = []
class_label = []
for r in new_pos:
    new_reviews.append([r, 1])
for r in new_neg:
    new_reviews.append([r, 0])

# shuffles the list
random.shuffle(new_reviews)

# creates training and test sets
X = [review[0] for review in new_reviews]
Y = [review[1] for review in new_reviews]

# use CountVectorizer
X = cv.transform(X)

# makes the prediction
predicted = nb.predict(X)
print('Accuracy for the new reviews:', accuracy_score(Y, predicted))
```
